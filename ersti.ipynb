```python
iris_melted = pd.melt(iris, id_vars = ['species'])
iris_melted.head()
```

```python
%matplotlib inline
sns.boxplot(x="variable", y="value", hue="species", data= iris_melted)
```

```python
sns.factorplot(x='species', y='value', col='variable', kind= 'violin', data= iris_melted)
```

```python
from scipy.stats import ttest_ind
setosa = iris[iris['species']=='setosa']
virginica = iris[iris['species']=='virginica']

print(ttest_ind(setosa['sepal_length'], virginica['sepal_length']))
print(ttest_ind(setosa['sepal_width'], virginica['sepal_width']))
```

```python
from nilearn import datasets
anatomical_image = datasets.load_mni152_template()
from nilearn import plotting
plotting.plot_anat(anatomical_image, title= "MNI TEMPLATE")
```

```python
localizer_dataset = datasets.fetch_localizer_button_task(n_subjects= 1, get_anats=True)
localizer_anat_filename = localizer_dataset.anats[0]
print(localizer_anat_filename)
localizer_tmap_filename = localizer_dataset.tmaps[0]
print(localizer_tmap_filename)
```

```python
plotting.plot_stat_map(localizer_tmap_filename, bg_img= localizer_anat_filename, threshold=3, title= 'Stattistical map' )
```

```python
plotting.plot_glass_brain(localizer_tmap_filename, threshold=3, plot_abs=False, colorbar=True, title= 'Stattistical map glass_brain' )
```

```python
plotting.plot_stat_brain(localizer_tmap_filename, threshold=3, display_mode= 'z', cut_coords=5, colorbar=True, title= 'Stattistical map glass_brain' )
```
